```python
import math

def create_board():
    return [' ' for _ in range(9)]

def print_board(board):
  for i in range(3):
      print('|'.join(board[i*3:(i+1)*3]))
      if i<2:
        print('-'*5)

def check_winner(board):
  win_combinations=[
      [0,1,2],[3,4,5],[6,7,8],
      [0,3,6],[1,4,7],[2,5,8],
      [0,4,8],[2,4,6]
  ]
  for combo in win_combinations:
     if board[combo[0]]==board[combo[1]]==board[combo[2]]!=' ':
       return board[combo[0]]
  if ' ' not in board:
     return 'T'
  return None

def minimax(board,is_maximizing,a,b):
  winner=check_winner(board)
  if winner=='O': return 1
  elif winner=='X': return -1
  elif winner=='T':return 0

  if is_maximizing:
    max_eval=-math.inf
    for i in range(9):
      if board[i]==' ':
        board[i]='O'
        eval=minimax(board,False,a,b)
        board[i]=' '
        max_eval=max(max_eval,eval)
        a=max(a,eval)
        if b<=a:
          break
    return max_eval
  else:
    min_eval=math.inf
    for i in range(9):
      if board[i]==' ':
        board[i]='X'
        eval=minimax(board,True,a,b)
        board[i]=' '
        min_eval=min(min_eval,eval)
        b=min(b,eval)
        if b<=a:
          break
    return min_eval

def aimove(board):
  bs=-math.inf
  move=None
  for i in range(9):
    if board[i]==' ':
      board[i]='O'
      score=minimax(board,False,-math.inf,math.inf)
      board[i]=' '
      if score>bs:
        bs=score
        move=i
  return move

def human(board):
  while True:
    try:
      pos=int(input("Enter your move from 1-9: "))-1
      if board[pos]==' ':
        return pos
      else:
        print("Already there. Try another.")
    except (ValueError,IndexError):
      print("Invalid number.Choose from 1-9")

def play():
  board=create_board()
  print("TIC-TAC-TOE")
  print_board(board)

  while True:
    pos=human(board)
    board[pos]='X'
    print_board(board)
    result=check_winner(board)
    if result:
      break
    print("AI is thinking")
    pos=aimove(board)
    board[pos]='O'
    print_board(board)
    result=check_winner(board)
    if result:
      break

  if result=='X':
    print("You Win")
  elif result=='O':
    print("AI Wins")
  else:
    print("Tie")

if __name__=='__main__':
  play()
```
