```python
import math
from IPython.display import display,clear_output
import ipywidgets as widgets

board=[' ']*9
buttons=[widgets.Button(description=' ',layout=widgets.Layout(width='60px',height='60px'))for _ in range(9)]

def display_board():
  clear_output(wait=True)
  grid=widgets.GridBox(buttons,layout=widgets.Layout(grid_template_columns="60px 60px 60px",grid_template_rows="60px 60px 60px",grid_gap="5px 5px"))
  display(grid)

def check_winner(board):
  wins=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
  for a,b,c in wins:
    if board[a]==board[b]==board[c] and board[a]!=' ':
      return board[a]
  if ' ' not in board:
    return 'T'
  return None

def minimax(board,is_max,alpha,beta):
  winner=check_winner(board)
  if winner=='O': return 1
  if winner=='X': return -1
  if winner=='T': return 0

  if is_max:
    best=-math.inf
    for i in range(9):
      if board[i]==' ':
        board[i]='O'
        val=minimax(board,False,alpha,beta)
        board[i]=' '
        best=max(best,val)
        alpha=max(alpha,best)
        if beta<=alpha:
          break
    return best
  else:
    best=math.inf
    for i in range(9):
      if board[i]==' ':
        board[i]='X'
        val=minimax(board,True,alpha,beta)
        board[i]=' '
        best=min(best,val)
        beta=min(beta,best)
        if beta<=alpha:
          break
  return best


def ai_move():
  best_score=-math.inf
  move=None
  for i in range(9):
    if board[i]==' ':
      board[i]='O'
      score=minimax(board,False,-math.inf,math.inf)
      board[i]=' '
      if score>best_score:
        best_score=score
        move=i
  return move


def on_click(i):
  def handler(btn):
    if board[i]!=' ' or check_winner(board):
      return
    board[i]='X'
    buttons[i].description='X'
    buttons[i].disabled=True
    winner=check_winner(board)
    display_board()
    if winner:
      show_result(winner)
      return

    ai=ai_move()
    if ai is not None:
      board[ai]='O'
      buttons[ai].description='O'
      buttons[ai].disabled=True
      winner=check_winner(board)
      display_board()
      if winner:
        show_result(winner)
  return handler

def show_result(result):
  if result=='T':
    print("It is a tie")
  else:
    print(f"{result} wins!")

for i in range(9):
  buttons[i].on_click(on_click(i))

print("You are X. Click a square to play.")
display_board()
```
